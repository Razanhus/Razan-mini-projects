```python
import pandas as pd
import numpy as np
import cv2
from ultralytics import YOLO
```

```python
# Load a YOLO model (replace 'yolov8n.pt' with the path to your model)
model = YOLO('yolov8n.pt')
```

```python
# Load the video
cap = cv2.VideoCapture('/content/istockphoto-500104950-640_adpp_is (2).mp4')
```

```python
# Create a VideoWriter object to save the output video
out = cv2.VideoWriter('output_yolo2.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30,
                      (int(cap.get(3)), int(cap.get(4))))
```

```python
while True:
    ret, img = cap.read()
    if not ret:
        break

    # Perform inference
    results = model(img)

    # Process results
    for result in results:
        for detection in result.boxes:  # Iterate over detected boxes
            x1, y1, x2, y2 = map(int, detection.xyxy[0])  # Get box coordinates
            conf = detection.conf[0]  # Confidence score
            cls = int(detection.cls[0])  # Class ID
            label = model.names[cls]  # Get class name

            if label in ["car", "truck"]:  # Filter for specific labels
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(img, f'{label} {conf:.2f}', (x1, y1 - 5),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the annotated frame to the output video
    out.write(img)
```

```python
# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
